merge join and concatenate

```python
df_products = pd.DataFrame({
    'product_id': [101, 102, 103, 104, 105],
    'product_name': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Webcam'],
    'price': [1200, 25, 75, 300, 50]
})
data_sales = {
    'sale_id': [1, 2, 3, 4, 5, 6],
    'product_id': [101, 103, 102, 106, 101, 107], # 106 and 107 are new products not in df_products
    'quantity': [1, 2, 1, 3, 1, 1],
    'sale_date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05', '2023-01-06']
}
df_sales = pd.DataFrame(data_sales)

# Save DataFrames to CSV files in Google Drive
df_products.to_csv('/content/drive/MyDrive/products.csv', index=False)
df_sales.to_csv('/content/drive/MyDrive/sales.csv', index=False)
```

```python
df_sales
```

```python
df_products
```

```python
df_products.merge(df_sales)
#inner join (checks similar columns and combines it)
```

```python
df_products.merge(df_sales,how='left')
```

```python
df_products.merge(df_sales,how='right')
#right join
```

```python
df_products.merge(df_sales,how='outer')
#outer join gives every thing of both dataframes
```

```python
df_products.merge(df_sales,how='cross')
#takes one rown and check with all the provided ones
```

```python
df_products.join(df_sales.set_index('product_id'),on='product_id')
#similar indexes can be combined together form join
#if both tables had diff ids then
#df.merge(df1, left_on='id', right_on='emp_id')
```

```python
df_products.set_index('product_id').join(df_sales.set_index('product_id'))
# similar to join on e.id=s.id on sql (can stil do how= 'left right etc') use merge cuz more simpler
```
